```python
import pandas as pd
import json
```

```python

def gfw2csv(fName):
    with open(fName) as f:
        d = json.load(f)
    data=d['entries']
    df = pd.DataFrame(data)
    df=pd.concat([df, df['position'].apply(pd.Series)], axis=1)
    df.to_csv('gfw.csv')
    return df
```

```python
gfw2csv('may24.json')
```
